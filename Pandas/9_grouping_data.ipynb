```python
def grouper(df,idx,col):
    if 80<=df[col].loc[idx]<=90:
        return "80-90"
    elif 50<=df[col].loc[idx]<=70:
        return "50-70"
    else:
        return "others"
```

```python
g=df.groupby(lambda idx: grouper(df,idx,"temperature"))
g
```

```python
for key,data in g:
    print("\n")
    print("key",key)
    print("data",data)
```
