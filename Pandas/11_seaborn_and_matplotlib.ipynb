```python
import pandas as pd
from matplotlib import pyplot as plt
df_plot=pd.read_excel("linechart.xlsx")
df_plot
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_plot["Quarter"],df_plot["Fridge"],color="red",label="Fridge")
plt.plot(df_plot["Quarter"],df_plot["Dishwasher"],color="grey",label="Dishwasher")
plt.plot(df_plot["Quarter"],df_plot["Washing Machine"],color="black",label="Washing Machine")
plt.title("Yearly Sales")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
total_sales=df_plot[["Fridge","Dishwasher","Washing Machine"]].sum()
print(total_sales)
print(type(total_sales))
```

```python
plt.pie(total_sales, labels=total_sales.index)
plt.show()
```

```python
plt.pie(total_sales, labels=total_sales.index,autopct='%1.2f%%')
plt.show()
```

```python
plt.pie(total_sales, labels=total_sales.index,autopct='%1.2f%%',explode=(0.1,0.3,0))
plt.show()
```

```python
plt.pie(total_sales, labels=total_sales.index,autopct='%1.2f%%',explode=(0.1,0.3,0),shadow=True,startangle=180)
plt.show()
```
