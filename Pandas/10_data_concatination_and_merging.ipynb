```python
import pandas as pd
df=pd.DataFrame({
    'city':['Dehli','Dehradun','Mumbai','Rajasthan'],
    'weather':['Cloudy','Sunny','Rainy','hot']
})
df
```

```python
df_1=pd.DataFrame({
    'city':['Kolkata','Hyedrabad','Banglore'],
    'weather':['Cloudy','Sunny','Rainy']
})
df_1
```

```python
df_2=pd.DataFrame({
    'temperature':[60,30,45],
    'pollution':['polluted','clear','foggy']
})
df_2
```

```python
pd.concat([df,df_1])
```

```python
pd.concat([df,df_1],ignore_index=True)
```

```python
pd.concat([df,df_1],keys=["north","south"])
```

```python
pd.concat([df,df_2],axis=1)
```

```python
df_c=pd.concat([df,df_1],ignore_index=True)
df_c
```

```python
df_3=pd.DataFrame({
    'city':['Dehradun','Dehli','Kolkata','Mumbai','Hyedrabad','Banglore','Madhya Pradesh'],
    'temperature':[30,50,45,40,33,44,50]
})
df_3
```

```python
pd.merge(df_c,df_3,on="city")     #by default *how=inner* 
```

```python
pd.merge(df_c,df_3,on="city",how="inner")   #only prints values common in both 
```

```python
pd.merge(df_c,df_3,on="city",how="left")   #prints values of left df and merges into them
```

```python
pd.merge(df_c,df_3,on="city",how="right")   #prints values of right df and merges into them
```

```python
pd.merge(df_c,df_3,on="city",how="outer")  #prints all values and merges 
```
