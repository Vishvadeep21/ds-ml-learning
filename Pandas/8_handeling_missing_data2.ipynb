```python
# df.replace() function is used to replace values of the dataframe
import pandas as pd
df=pd.read_csv("weather_data.csv")
df
```

```python
# Way 01
df_1=df.replace(-99999,0)
df_1
```

```python
# way 02
import numpy as np
df_2=df.replace([-99999,-88888],np.nan)
df_2
```

```python
# way 03
df_3=df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':'no event'
},np.nan)
df_3
```

```python
# way 04
df_4=df.replace({
    -99999:np.nan,
    -88888:np.nan,
    'no event':'sunny'
})
df_4
```
