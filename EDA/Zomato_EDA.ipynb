```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib 
matplotlib.rcParams['figure.figsize'] = (12, 6)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
plt.show()
```

```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country
```

```python
df_final=pd.merge(df,df_country, on='Country Code', how='left')
df_final.head()
```

Find top 3 Countries that uses zomato


```python
df_final.columns
```

```python
country_index=df_final.Country.value_counts().index
country_value=df_final.Country.value_counts().values
```

```python
plt.pie(country_value[:3],labels=country_index[:3],autopct="%1.2f%%")
plt.show()
```

observation:
top 3 countries are 
India
US
UK

```python
df_final.columns
```

```python
rating=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
rating
```

```python
sns.barplot(x="Aggregate rating",y="Rating counts",data=rating,hue="Rating color", palette=['black','red','orange','yellow','green','green'])
plt.show()
```

```python
sns.countplot(data=rating,x='Rating color',hue='Rating color',palette=['black','red','orange','yellow','green','green'])
plt.show()
```
